```python
df_segment_category = df.groupby(['segment', 'category']).agg({
    'sales' : 'sum'
})
df_segment_category = df_segment_category.reset_index()
df_segment_category.head()
```

```python
#1.Group Bar Chart
bar1 = go.Bar(
    x =df_segment_category[df_segment_category.segment == 'Corporate'].category,
    y =df_segment_category[df_segment_category.segment == 'Corporate'].sales,
    name = 'Corporate'
)

bar2 = go.Bar(
    x =df_segment_category[df_segment_category.segment == 'Consumer'].category,
    y =df_segment_category[df_segment_category.segment == 'Consumer'].sales,
    name = 'Consumer'
)

bar3 = go.Bar(
    x =df_segment_category[df_segment_category.segment == 'Home Office'].category,
    y =df_segment_category[df_segment_category.segment == 'Home Office'].sales,
    name = 'Home Office'
)

layout={
    'xaxis':{
        'title':'Category',
    },
    'yaxis':{
        'title':'Sales',
    }
}

data = [bar1, bar2, bar3]
fig1 = go.Figure(data = data, layout = layout)
iplot(fig1)
```

```python
df_segment_category1 = df.groupby(['segment', 'category']).agg({
    'quantity' : 'sum'
})
df_segment_category1 = df_segment_category1.reset_index()
df_segment_category1.head()
```

```python
#1.Group Bar Chart
bar4 = go.Bar(
    x =df_segment_category1[df_segment_category1.segment == 'Corporate'].category,
    y =df_segment_category1[df_segment_category1.segment == 'Corporate'].quantity,
    name = 'Corporate'
)

bar5 = go.Bar(
    x =df_segment_category1[df_segment_category1.segment == 'Consumer'].category,
    y =df_segment_category1[df_segment_category1.segment == 'Consumer'].quantity,
    name = 'Consumer'
)

bar6 = go.Bar(
    x =df_segment_category1[df_segment_category1.segment == 'Home Office'].category,
    y =df_segment_category1[df_segment_category1.segment == 'Home Office'].quantity,
    name = 'Home Office'
)

layout={
    'title':{
        'text':'Buy Quantity',
        'x' : 0.47,  
    },
    'xaxis':{
        'title':'Category',
    },
    'yaxis':{
        'title':'Quantity',
    }
}

data = [bar4, bar5, bar6]
fig1 = go.Figure(data = data, layout = layout)
iplot(fig1)
```

```python
#2.Stack Bar Chart
data = [bar1, bar2, bar3]
layout = {
    'barmode':'stack',
}

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
